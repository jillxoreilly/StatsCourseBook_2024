### Control Variables

Which do you think might be important controls here? 

There is no right or wrong answer here but think about your reasons for selecting your control variables (don’t just throw all of them in!).

Specify two regression models - Model 1 includes just the main explanatory variable. Model 2 adds the control variables of your choice (and keeps the main explanatory variable). Calculate the RMSE for both. 


<font color=red>For my first model I will just include `ladder`</font>

```python
# Model 1
# first we run this line to tell statsmodels where to find the data and the explanatory variables
reg_formula = sm.regression.linear_model.OLS.from_formula(data = data, formula = 'happy ~ ladder')

# then we run this line to fit the regression (work out the values of intercept and slope)
# the output is a structure which we will call reg_results
reg_results_model1 = reg_formula.fit()

# let's view a summary of the regression results
reg_results_model1.summary() 
```

<font color=red>Now to include some control variables</font>

```python
# Model 2
# first we run this line to tell statsmodels where to find the data and the explanatory variables
reg_formula = sm.regression.linear_model.OLS.from_formula(data = data, 
                    formula = 'happy ~ ladder + age + income + marital')

# then we run this line to fit the regression (work out the values of intercept and slope)
# the output is a structure which we will call reg_results
reg_results_model2 = reg_formula.fit()

# let's view a summary of the regression results
reg_results_model2.summary() 
```

### Compare your two models:

* Which is better fitting in terms of the $R^2$? 
* And which has a smaller spread of values around the regression line? 

<font color=red> The R2 is higher for the model with control variables (r2 = 0.147) vs the basic model (r2=0.121)

<font color=red> The spread of points around the regression line is determined by the SSE </font>

```python
print(reg_results_model1.ssr)
print(reg_results_model2.ssr)
```

<font color=red> The spread is lower for the model with control variables.
    
<font color=red> Obviously the model with control variables has to be a better fit as the model is fit to the same data (no cross validation etc)</font>
